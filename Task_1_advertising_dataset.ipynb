```python
#Maximum and minimum money spent on each platform

print("Money spent on TV platform: Maximum = ",df['TV'].max()," Minimum = ",df['TV'].min())

print("Money spent on Radio platform: Maximum = ",df['Radio'].max()," Minimum = ",df['Radio'].min())

print("Money spent on Newspaper platform: Maximum = ",df['Newspaper'].max()," Minimum = ",df['Newspaper'].min())
```

```python
# Visualization Maximum and minimum money spent on each platform 

df_1 = pd.DataFrame({
    'Platform': ['TV', 'Radio','Newspaper'],
    'Max_money_spent': [df['TV'].max(), df['Radio'].max(),df['Newspaper'].max()],
    'Min_money_spent': [df['TV'].min(), df['Radio'].min(),df['Newspaper'].min()]
})
fig, ax1 = plt.subplots(figsize=(10, 10))
tidy = df_1.melt(id_vars='Platform').rename(columns=str.title)
sns.barplot(x='Platform', y='Value', hue='Variable', data=tidy, ax=ax1)
ax1.bar_label(ax1.containers[0])
ax1.bar_label(ax1.containers[1])
plt.ylabel('Money')
plt.title('Max. and Min money spent on each Platform',fontsize=15)
```

```python
# Average amount of money spent on each platform

print("Average money spent on TV platform = ",df['TV'].mean())

print("Average money spent on Radio platform = ",df['Radio'].mean())

print("Average money spent on Newspaper platform = ",df['Newspaper'].mean())
```

```python
# Visualization Average amount of money spent on each platform

df_2 = pd.DataFrame({
    'Platform': ['TV', 'Radio','Newspaper'],
    'Average_money_spent': [df['TV'].mean(), df['Radio'].mean(),df['Newspaper'].mean()],
    
})
fig, ax1 = plt.subplots(figsize=(10, 10))
tidy = df_2.melt(id_vars='Platform').rename(columns=str.title)
sns.barplot(x='Platform', y='Value', hue='Variable', data=tidy, ax=ax1)
ax1.bar_label(ax1.containers[0])
plt.ylabel('Money')
plt.title('Average money spent on each Platform',fontsize=15)
```

```python
data = [df['TV'].sum(), df['Radio'].sum(), df['Newspaper'].sum()]
labels = ['TV', 'Radio', 'Newspaper']

#define Seaborn color palette to use
colors = sns.color_palette('pastel')[0:5]

#create pie chart
plt.figure(figsize=(10,10))
plt.pie(data, labels = labels, colors = colors, autopct='%.0f%%')
plt.title('Money spent on each Platform',fontsize=20)
```

### Linear Regression Model:

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Sales'], axis = 1),
                                                   df['Sales'],test_size = 0.20,
                                                   random_state = 12)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
accuracy = model.score(x_test, y_test)
print(accuracy*100, '%')
```
